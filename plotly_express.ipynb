## plotly express

```python
import plotly.express as px
```

```python
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species")
fig.show()
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", marginal_y="violin",
                 marginal_x = "box", trendline="ols", template="simple_white")
fig.show()
```

```python
import plotly.express as px
df2 = px.data.tips()
fig = px.bar(df2, x="sex", y="total_bill", color="smoker", barmode="group")
fig.show()
```

```python
df3 = px.data.gapminder()
fig = px.scatter(df3, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
                 size="pop", color="continent", hover_name="country", facet_col="continent",
                 log_x = True, size_max = 45, range_x = [100, 100000], range_y = [25,90])             
                

fig.show()
```

## gráficos alternativos 

```python
df5 = px.data.gapminder().query("year == 2007")
fig = px.sunburst(df5, path = ["continent", "country"], values="pop",
                  color = "lifeExp", hover_data=["iso_alpha"])

fig.update_layout(height=500)

fig.show()
```
